```python
#datasets
diamonds_train = pd.read_csv('../db/diamonds_train.csv')
diamonds_test = pd.read_csv('../db/diamonds_test.csv')
```

```python
#review the datatype
diamonds_train.info()
```

```python
#reviewing unique values in categorical features
print("Cut: ",set(diamonds_train["cut"]))
print("Color: ",set(diamonds_train["color"]))
print("Clarity: ",set(diamonds_train["clarity"]))
```

```python
#To formulate relationships with the price of diamond we must have a standard reference in which we can compare values.
#I created a new column comparing the prices per unit carat, and this will be the one I will use to compare the other features.
diamonds_train['price/wt']=diamonds_train['price']/diamonds_train['carat']
print(diamonds_train.groupby('cut')['price/wt'].mean().sort_values())
print(diamonds_train.groupby('color')['price/wt'].mean().sort_values())
print(diamonds_train.groupby('clarity')['price/wt'].mean().sort_values())
diamonds_train = diamonds_train.drop(['price/wt','table'], axis=1)
```

```python
#mapping categorical features with numbers, in order to convert to numerical.
diamonds_train['cut']=diamonds_train['cut'].map({'Ideal':5,'Good':2,'Very Good':3,'Fair':1,'Premium':4})
diamonds_train['color']=diamonds_train['color'].map({'E':2,'D':1,'F':3,'G':4,'H':5,'I':6,'J':7})
diamonds_train['clarity']=diamonds_train['clarity'].map({'VVS1':7,'IF':8,'VVS2':6,'VS1':5,'I1':1,'VS2':4,'SI1':3,'SI2':2})
```

```python
#Correlation Matrix
diamonds_corr = diamonds_train.corr()
diamonds_corr
```

```python
#heatmap to review the correlation in an easier way
sns.set(rc={'figure.figsize':(12,8)})
sns.heatmap(diamonds_corr, annot=True);
plt. savefig("..\images\image1.png", dpi=300)
```

```python
#Although, ‚Äúcut‚Äù, ‚Äúcolor‚Äù and ‚Äúclarity‚Äù features have a low correlation score, 
#I prefer to keep them as I do believe they play a role in determining the price of a diamond.
#if we study the correlation matrix, it‚Äôs clear that the price and carat i.e. weight of a diamond has a very good correlation score.
#So, I inserted new features like cut per weight score and drop the features like ‚Äúcut‚Äù, ‚Äúcolor‚Äù and ‚Äúclarity‚Äù.
diamonds_train['cut/wt']=diamonds_train['cut']/diamonds_train['carat']
diamonds_train['color/wt']=diamonds_train['color']/diamonds_train['carat']
diamonds_train['clarity/wt']=diamonds_train['clarity']/diamonds_train['carat']
diamonds_train = diamonds_train.drop(['cut','color','clarity','depth','city','x','y','z'], axis=1)
```

```python
#I had already tested using RandomForestRegressor(), but I want to improve my result by optimizating the hyper-parameters,
#so I will use RandomizedSearchCV. to get the best options I  

mdl = RandomForestRegressor()

mdl_random = RandomizedSearchCV(estimator = mdl, param_distributions = {'n_estimators': [16, 32, 64, 128, 256, 512],
              'max_depth': [2, 4, 8, 16]},
                               cv = 5, 
                               verbose=3, 
                               random_state=42, 
                               n_jobs = -1)
print ("Training model...")
# Fit the random search model
mdl_random.fit(X,Y)
print ("Model trained.")
```
